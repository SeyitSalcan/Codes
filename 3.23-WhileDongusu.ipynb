```python
x = 0  # 0 dan başlayarak 1 artıtak ekrana yazdırma while döngüsü ile
while x < 10:
    print(x)
    x = x + 1
```

```python
benimListem = [1,2,3,4,5] #liste ataması
```

```python
benimListem.pop() # listenin pop elemanonın atılması
```

```python
benimListem #listeden atılmış hali
```

```python
benimListem.append(5) # 5 elemanı ekleme
```

```python
while 3 in benimListem: # 3 elemanı listeden çıkarılana kadar döngü devam eder.
    print("3 hala listenin içerisinde")
    benimListem.pop()
```

```python
numara = 0
```

```python
while numara < 5: # numara değişkenini bir bir artırarak 4 olduğunda döngü sonlandırma
    if numara == 4:
        break
    print(numara)
    numara = numara + 1
```
