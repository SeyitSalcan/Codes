## enumerate

```python
index = 0 # for döngüsü ve indeks ** değer yazılması
for numara in list(range(5,15)):
    print(f"güncel numara: {numara} güncel index: {index}")
    index = index + 1
```

```python
for eleman in enumerate(list(range(5,15))): #enumerate ile indeks -- değer birlikte yazılması
    print(eleman)
```

```python
for (index,numara) in enumerate(list(range(5,15))): # indeks ve değer örneği
    print(numara)
```

## zip

```python
yemekListesi = ["muz","ananas","elma"]
kaloriListesi = [100,200,300]
gunListesi = ["pazartesi","salı","çarşamba"]
```

```python
ziplenmisListe = list(zip(yemekListesi,kaloriListesi,gunListesi)) # zipleyerek listelerin tuple dönüşmesi
```

```python
for eleman in ziplenmisListe:
    print(type(eleman)) #typelerinin yazılması
```

```python
ziplenmisListe # zipli liste
```
