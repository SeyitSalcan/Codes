```python
benimListem = [10,20,30,40,50] # liste ataması
```

```python
if 10 in benimListem:
    print("evet içerisinde mevcut") # liste içerisindeki eleman araması
```

```python
benimListem[0] * 5 / 3 #listenin 0.indexi matematiksel işlem
```

```python
benimListem[1] * 5 / 3 #listenin 1.indexi matematiksel işlem
```

```python
print("döngü başladı")  # for döngüsü tanıtımı.. listedki her elemeanın 5 ile çarpılıp 3 bölünmesi
for numara in benimListem:
    print(numara * 5 / 3)
print("döngü bitti")
```

```python
for num in benimListem:  # listedek iher eleamnın mateamtik işlemi sonrası değişkene ataması
    yeniNumara = num * 5 / 3
    print(yeniNumara)
```

```python
yeniListe = [1,2,3,4,5,6] # yeniliste ataması
```

```python
for rakam in yeniListe: # yeniListe mode işlemi
    if rakam % 2 == 0:
        print(rakam)
```

```python
yeniString = "Atıl Samancıoğlu" #string for döngüsü
```

```python
for harf in yeniString:
    print(harf)
```

```python
benimTuple = (1,2,3,4,5) #tuple atamsı
```

```python
for eleman in benimTuple: # tuple for işlemleri
    print(eleman - 10)
```

```python
koordinatListesi = [(10.2,15.2),(32.4,16.2),(40.2,20.2)] # tuple listesi
```

```python
type(koordinatListesi[0]) #type of tuple
```

```python
for eleman in koordinatListesi: # tuple listesinin for döngüsü 
    print(eleman)
```

```python
for (x,y) in koordinatListesi: # tuple listesini ikili yazılması
    print(y)
```

```python
benimGaripListem = [(1,2,3),(4,5,6),(7,8,9)] # uclu tuple listesi
```

```python
for (x, y, z) in benimGaripListem: # tuple listesinden znin çıktı olarak yazılması
    print(z)
```
