## input & return

```python
def merhabaDunya(yazdirilacakIsim): # veri alan fonksiyon tanıtımı
    print("merhaba")
    print(yazdirilacakIsim)
```

```python
merhabaDunya("Atıl")
```

```python
merhabaDunya("python")
```

```python
def merhaba(isim = "atıl"): # ilk değeri verilen veri alan fonksiyon
    print("merhaba")
    print(isim)
```

```python
merhaba()
```

```python
merhaba("python")
```

```python
def toplama(numara1, numara2): # int deger alan fonskyionalr
    sonuc = numara1 + numara2
    print(sonuc)
```

```python
toplama(10,20)
```

```python
toplama(-200,350)
```

```python
def superToplama(num1, num2, num3): # ikşden fazla değer alan fonksiyon
    print(num1+num2+num3)
```

```python
superToplama(10,20,30)
```

```python
yeniDegisken = toplama(10,20) # değişkene atama
```

```python
def kontrolFonksiyon(s): # kontrol yapılı fonk.
    if s == "atıl":
        print("verdiğiniz string atıl")
    else:
        print("verdiğiniz string başka bir şey")
```

```python
kontrolFonksiyon("atıl")
```

```python
kontrolFonksiyon("python")
```

## args & kwargs

```python
def yeniToplama(*args): # pointer ile fonk. tanımı
    return sum(args)
```

```python
yeniToplama(10,20,30,40,50,60)
```

```python
yeniToplama(10,20)
```

```python
def benimFonksiyonum(*args):
    return args
```

```python
type(benimFonksiyonum(20,30,40))
```

```python
def ornekFonksiyon(**kwargs): 
    return(kwargs)
```

```python
type(ornekFonksiyon(muz = 100, elma = 200, ananas = 300))
```

```python
def keyWordKontrolu(**kwargs):
    if "atıl" in kwargs:
        print("atıl var")
    else:
        print("atıl yok")
```

```python
keyWordKontrolu(ahmet = 70, zeynep = 50, mehmet = 40)
```
